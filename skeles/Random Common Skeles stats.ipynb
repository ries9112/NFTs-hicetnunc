```python
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt

from IPython.core.display import display, HTML
from IPython.core.display import Markdown as md
display(HTML("<style>.container { width:85% !important; }</style>"))

np.random.seed(42)

pd.set_option('display.max_columns', None)
plt.style.use('fivethirtyeight')

f38_palette = [
    "#30a2da",
    "#fc4f30",
    "#e5ae38",
    "#6d904f",
    "#8b8b8b",
]

sk_full = pd.read_csv("~/workspace/rcs/skeles_full.csv")
sk_full.head()
```

```python
trait_cols = ['hat', 'glasses', 'body', 'arm', 'hand', 'bottom', 'top', 'background']
```

```python
import math
from functools import reduce
from collections import Counter


col_counters = {}
for c in trait_cols:
    cntr = Counter(sk_full.loc[:, c].values)
    col_counters[c] = cntr
```

```python
total = len(sk_full)
def get_trait_perc(x):
    percs = []
    for c in x.index.values:
        percs.append(col_counters[c][x[c]] / total)
    return percs

sk_full.loc[:, "percs"] = (
    sk_full.loc[:, trait_cols].apply(get_trait_perc, axis=1)
)
sk_full.loc[:, "rarity"] = sk_full.percs.apply(lambda x: reduce(lambda x, y: x * y, x))
sk_full.loc[:, "n_with"] = list(map(lambda x: math.ceil(x), sk_full.rarity * len(sk_full)))
```

```python
from collections import Counter
```

```python
def n_set(x):
    c = Counter(x)
    mc = c.most_common()
    if mc[0][1] == 1:
        return 1
    to_ret = 0
    fact = 1
    for item, val in mc:
        if val <= 1:
            return to_ret
        else:
            to_ret += fact * val
            fact *= 10
    return to_ret
sk_full.loc[:, "multiple"] = sk_full.filter(trait_cols).apply(n_set, axis=1)
```

```python
def map_for_ids(df, f):
    subset = df[f].filter(["token_id", "rarity", "n_with"])
    to_md = list(
        map(lambda x: f"[{x[0]}](https://objkt.com/asset/rcs/{x[0]}) -- {x[1]} -- {x[2]}",
            subset.values
        )
    )

    display(md(
        str("<p></p>".join(to_md))
    ))
map_for_ids(sk_full, (sk_full.multiple == 3))
```
